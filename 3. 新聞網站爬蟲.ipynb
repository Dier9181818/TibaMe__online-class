```python
import requests
from bs4 import BeautifulSoup
import re


new_list = []
url = 'https://money.udn.com/money/index'

## 1.填寫headers
headers = {'User-Agent':'Mozilla/5.0 (Windows NT 6.1; Win64; x64; rv:57.0) Gecko/20100101 Firefox/57.0'}

response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, 'html.parser')
# print(response.text)


## 2.抓取即時區域所有新聞title以及新聞內容連結link，並將每篇新聞的連結link append到new_list內

div_title = soup.select('div[class="story__list"]')
soup2 = BeautifulSoup(str(div_title), 'html.parser')
a_tags = soup2.select('a')

for i in a_tags:
    title = i.text.replace(' ','').replace('\n','')
    dw_link = i['href'].replace('\n','')
    print('title: {}, link: {}'.format(title, dw_link))
    new_list.append(dw_link)
```

```python
from time import sleep
import os
from bs4 import BeautifulSoup
from urllib.request import urlretrieve


def createFolder(directory):
    try:
        if not os.path.exists(directory):
            os.makedirs(directory)
    except OSError:
        print ('Error: Creating directory. ' +  directory)


def save_news(url):
    
   
    ## 3.填寫headers
    headers = {'User-Agent':'Mozilla/5.0 (Windows NT 6.1; Win64; x64; rv:57.0) Gecko/20100101 Firefox/57.0'}
    response = requests.get(url, headers=headers)
    sleep(1)

    # ====================== Title ======================

    soup = BeautifulSoup(response.text, 'html.parser')

    ## 4. 使用find_all去抓取每篇新聞的title(指定是哪個tag以及指定id)，將結果所產生之list放到new_title變數內
    new_title = soup.find_all('h1', {'id':'story_art_title'})

    save_path = './news/'+new_title[0].text.replace('/', '_')
    createFolder(save_path)  

    article_body = soup.find_all('section', {'id':'article_body'})
    print('new title: {}'.format(new_title[0].text.replace('/', '_').replace(':', '')))


    # ====================== Content ======================

    soup2 = BeautifulSoup(str(article_body[0]), 'html.parser')

    with open(save_path+'/contents.txt', 'w', encoding='utf-8') as f:  
        ## 5. 抓取文章內p tags，並將新聞內文寫入contents.txt檔案內
        p_tags = soup2.find_all('p')

        print('new content: ')
        for i in p_tags:
            print(i.text)
            f.write(i.text)


    # ====================== img ======================       
    
    ## 6. 抓取文章內所有img tags，並將圖片下載並存入對應新聞資料夾下
    img_tags = soup2.find_all('img')

    print('save image: ')
    for i in img_tags:
        print(i['src'])
        urlretrieve(i['src'], save_path+'/'+i['title'].replace('/', '_')+'.jpg')
        
    print('==========\n')
    

for new_url in new_list:
    print(new_url)
    save_news(new_url)
```
